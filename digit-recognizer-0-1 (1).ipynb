```python
import numpy as np 
import pandas as pd 
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = '/kaggle/input/mnist-handwritten-digits-0-and-1-dataset/Mnist_Binary_0_vs_1_1000_20x20.csv'
df = pd.read_csv(path)
df.head()
```

```python
X=df.drop("label",axis=1).to_numpy()
y=df["label"].to_numpy()
```

```python
print(X.shape,
y.shape
     )
```

```python
m,n=X.shape
fig,axes=plt.subplots(8,8,figsize=(8,8))
fig.tight_layout(pad=0.1)
for i,ax in enumerate(axes.flat):
    random_index=np.random.randint(m)
    X_random_reshape=X[random_index].reshape((20,20)).T
    ax.imshow(X_random_reshape,cmap="gray")
    ax.set_title(y[random_index])
    ax.set_axis_off()
    
```

```python
model = Sequential([
    tf.keras.Input(shape=(400,)),
    Dense(25,activation="sigmoid"),
    Dense(15,activation="sigmoid"),
    Dense(1,activation="sigmoid")
],name="my_model1"
                  )
```

```python
model.summary()
```

```python
L1_nums=400*25+25
L2_nums=25*15*15
L3_nums=15*1+1
```

```python
[layer1,layer2,layer3]=model.layers
```

```python
W1,b1=layer1.get_weights()
W2,b2=layer2.get_weights()
W3,b3=layer3.get_weights()
print(f"W1 shape = {W1.shape}, b1 shape = {b1.shape}")
print(f"W2 shape = {W2.shape}, b2 shape = {b2.shape}")
print(f"W3 shape = {W3.shape}, b3 shape = {b3.shape}")
```

```python
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(0.001)
)
```

```python
model.fit(X,y,epochs=100)
```

```python
prediction=model.predict(X[0].reshape(1,400))
print(f"predicting a zero={prediction}")
prediction_class=np.argmax(prediction,axis=1)[0]
print(prediction_class)
```
